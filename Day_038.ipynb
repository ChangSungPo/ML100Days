## 作業
在鐵達尼資料集中，
透過進階補值方法 KNN，將鐵達尼號中的 Titanic_train
.csv 中的 age 的遺失值進行補值。    
* step1: 觀察 Age 和 Pclass 與 Sex 是否有關連性?
* step2: 如果有關連性，運用 KNN ，取出 Age 、 Pclass、 Sex的資料，以 Sex 與 Pclass 補 Age 遺失值。 

```python
# import library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
from IPython.display import display
```

## 讀入資料

```python
train = pd.read_csv(".\Titanic_train.csv")
train
```

## 觀察 Age

```python
g = sns.catplot( x ="Sex" , y = "Age" , hue = "Pclass" ,
                data = train, kind = "box")
```

```python
#離散轉連續型資料
train_copy  = train.copy()#複製一份資料給 data_copy
#轉型 
#透過 preprocessing 轉換
from sklearn import preprocessing
#下面兩行程式，讓 Male = 1, FeMale = 0.
le = preprocessing.LabelEncoder()
train_copy["Sex"] = le.fit_transform(train_copy["Sex"])
train_copy  
```

```python
#計算資料點的倆倆距離
from sklearn.metrics.pairwise import nan_euclidean_distances
x = ["Age" , "Pclass", "Sex"] 
y = train_copy[x]
pd.DataFrame(nan_euclidean_distances(y))
```

```python
from sklearn.impute import KNNImputer
value_neighbors = 3
imputer = KNNImputer(n_neighbors=value_neighbors)
df_filled = pd.DataFrame(imputer.fit_transform(y))
df_filled 
```
